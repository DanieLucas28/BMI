Dentre os modelos disponíves, existem os que são mais comumente utilizados baseando-se no comportamento dos dados com o passar do tempo, incluindo ou não tendência e sazonalidade.

O modelo escolhido foi o **decomposição clássica**, pois permite que consideremos tendência e sazonalidade. Além disso, como as compras são feitas diariamente, um modelo que consiga se adequar bem a grupos de dias separadamente se fez necessário:
image.png



> Decomposição clássica



```python
#Importando as funções que serão utilizadas no modelo

from pandas import DataFrame, read_excel
from scipy.stats import linregress
from datetime import datetime
from datetime import timedelta
```

```python
path = "query.xlsx"
```

```python
#Calcular interceção e inclinação usando a função linregress do scipy


#Como a decomposição clássica requer uma quantidade grande de dados, limitei a aplicação a pelo menos 31 dados

data = read_excel(path)

if len(data) > 30:

    x = list(range(1, len(data) + 1, 1))
    y = data['quantidade'].tolist()
        
    slope, intercept, r, p, se = linregress(x, y)

data.head()
```

```python
#Cálculo da tendência linear, basicamente, y=ax+b

data['weekday'] = data['weekday'].str.strip()
data['linear_trend'] = (data.index + 1) * slope + intercept
mean = data['quantidade'].mean()

data.head()
```

```python
#Os dados agrupados por dia e somados

weekday = data.groupby(['weekday']).agg({'quantidade': ["sum","count"]}).reset_index()

#É calculada a média por dia

weekday['mean_day'] = weekday['quantidade']['sum'] / weekday['quantidade']['count']

weekday.head(7)
```

```python
#Média por dia dividida pela média de todos os dias (retorna o "peso por dia")

df = DataFrame()
df['mean_all'] = weekday['mean_day'] / mean
df.set_index(weekday['weekday'], inplace=True)

df.head(7)
```

```python
#O valor da encontrado na equação é multiplicado pelo "peso do dia"

data['trend_seasonality'] = data['linear_trend'] * df.loc[data['weekday'], 'mean_all'].tolist()
data.head()
```

```python
#cálculo da diferença entre o esperado e o encontrado no modelo

data['absolute_difference'] = abs(data['quantidade'] - data['trend_seasonality'])

data.head()
```

```python
#Cálculo da diferença entre o esperado e o encontrado agrupado por dia para determinar os intervalos considerando os erros

weekday2 = data.groupby(['weekday']).agg({'absolute_difference': ["sum", "count"]})
weekday2['difference_day'] = weekday2['absolute_difference']['sum'] / weekday2['absolute_difference']['count']

weekday2.head(7)
```

```python
#adição do intervalo considerando o erro para mais e para menos

data['lower_interval'] = data['trend_seasonality'] - weekday2.loc[data['weekday'], 'difference_day'].tolist()
data['upper_interval'] = data['trend_seasonality'] + weekday2.loc[data['weekday'], 'difference_day'].tolist()

data.head()
```

```python
#Cálculo da previsão para os próximos 7 dias úteis, considerando o intervalo superior

forecast = DataFrame(columns=['weekday', 'day', 'forecast'])
last_day = data.iloc[-1,0]
    
count=0
for day in range(len(data), len(data)+7):
    count +=1 
    forecast = forecast.append({'weekday': last_day + timedelta(count), 'day': (last_day + timedelta(count)).strftime('%A'), 'forecast': (slope*day+intercept)*df.loc[(last_day+timedelta(count)).strftime('%A'), 'mean_all']}, ignore_index=True)
    
forecast['upper_interval'] = forecast['forecast'] + weekday2.loc[forecast['day'], 'difference_day'].tolist()

forecast.head(7)
```
